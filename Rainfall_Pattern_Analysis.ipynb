## Load the data

### Subtask:
Load the `/content/rainfall in india 1901-2015.csv` file into a pandas DataFrame.


**Reasoning**:
The first step is to load the data into a pandas DataFrame using the provided file path.



```python
import pandas as pd

df = pd.read_csv('/content/rainfall in india 1901-2015.csv')
```

**Reasoning**:
Display the first few rows of the DataFrame to verify that the data has been loaded correctly.



```python
display(df.head())
```

## Initial data inspection

### Subtask:
Display the first few rows, check the data types, and look for missing values.


**Reasoning**:
Display the first few rows, check data types, and check for missing values as per the instructions.



```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

## Data cleaning

### Subtask:
Handle missing values, if any.


**Reasoning**:
Fill the missing values in the 'ANNUAL' and 'Oct-Dec' columns with the mean of their respective non-missing values and then verify the changes.



```python
df['ANNUAL'].fillna(df['ANNUAL'].mean(), inplace=True)
df['Oct-Dec'].fillna(df['Oct-Dec'].mean(), inplace=True)
display(df.isnull().sum())
```

**Reasoning**:
The previous step successfully filled the missing values in 'ANNUAL' and 'Oct-Dec' columns. Now, address the remaining columns with missing values by filling them with their respective means and then verify that all specified columns no longer have missing values.



```python
for col in ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'Jan-Feb', 'Mar-May', 'Jun-Sep']:
    df[col].fillna(df[col].mean(), inplace=True)
display(df.isnull().sum())
```

## Descriptive statistics

### Subtask:
Calculate and display descriptive statistics for the relevant columns.


**Reasoning**:
Select the relevant rainfall columns and calculate descriptive statistics using the .describe() method, then display the results.



```python
rainfall_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL', 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']
descriptive_stats = df[rainfall_cols].describe()
display(descriptive_stats)
```

## Trend analysis

### Subtask:
Analyze trends in rainfall over time.


**Reasoning**:
Group the DataFrame by 'YEAR', calculate the mean for each year, reset the index, and display the result.



```python
yearly_rainfall = df.groupby('YEAR').mean()
yearly_rainfall = yearly_rainfall.reset_index()
display(yearly_rainfall.head())
```

**Reasoning**:
The previous command failed because the 'SUBDIVISION' column is of object type and cannot be included in the mean calculation. I will select only the numeric columns before grouping.



```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
yearly_rainfall = df.groupby('YEAR')[numeric_cols].mean()
yearly_rainfall = yearly_rainfall.reset_index()
display(yearly_rainfall.head())
```

**Reasoning**:
The error indicates that 'YEAR' is already a column after grouping and calculating the mean. I need to remove 'YEAR' from the list of numeric columns before grouping, as it's the column I'm grouping by.



```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols.remove('YEAR')
yearly_rainfall = df.groupby('YEAR')[numeric_cols].mean()
yearly_rainfall = yearly_rainfall.reset_index()
display(yearly_rainfall.head())
```

## Correlation analysis

### Subtask:
Analyze the correlation between different columns, if applicable.


**Reasoning**:
Calculate and display the correlation matrix for the yearly_rainfall DataFrame to understand the relationships between different rainfall measurements.



```python
correlation_matrix = yearly_rainfall.corr()
display(correlation_matrix)
```

## Visualization

### Subtask:
Create charts and visualizations to represent the data and findings.


**Reasoning**:
Create a heatmap of the correlation matrix and a line plot of the annual rainfall trend over the years.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Yearly Rainfall Data')
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(yearly_rainfall['YEAR'], yearly_rainfall['ANNUAL'])
plt.title('Annual Rainfall Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.grid(True)
plt.show()
```
